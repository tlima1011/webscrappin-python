```python
try: 
    html = urlopen(url)
except HTTPError as e: 
    print(e)
except URLError as e: 
    print('The server could not be found')
```

```python
import requests
```

```python
response = requests.get(url, headers=headers)
```

```python
print(response.text)
```

```python
soup = BeautifulSoup(response.content, features="lxml")
```

```python
items = soup.find('div', class_='s-main-slot s-result-list s-search-results sg-row')
print(items)
```

```python
product =[]
prices = []
for i in soup.find_all('div', class_='a-size-base-plus a-color-base a-text-normal'):
    print(i.find('span', class_='a-size-base-plus a-color-base a-text-normal').next_sibling)
    if i.get('a-size-base-plus a-color-base a-text-normal') and i.get('a-price-whole'): 
        product.append([i.text for i in soup.find_all(class_='a-size-base-plus a-color-base a-text-normal')])
        prices.append([i.text for i in soup.find_all(class_='a-price-whole')])
```
